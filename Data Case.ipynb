```r
gapminder%>%
    filter(year==1957)%>%
    summarize(medianLifeExp=median(lifeExp),maxGdpPercap=max(gdpPercap))
```

```r
gapminder%>%
    group_by(year)%>%
    summarize(medianLifeExp=median(lifeExp),maxGdpPercap=max(gdpPercap))
```

```r
gapminder%>%
    filter(year==1957)%>%
    group_by(continent)%>%
    summarize(medianLifeExp=median(lifeExp),maxGdpPercap=max(gdpPercap))
```

```r
gapminder%>%
    filter(year==2007)%>%
    group_by(continent)%>%
    summarize(medianLifeExp=median(lifeExp),maxGdpPercap=max(gdpPercap))
```

```r
gapminder%>%
    group_by(continent,year)%>%
    summarize(medianLifeExp=median(lifeExp),maxGdpPercap=max(gdpPercap))
```

```r
by_year<-gapminder%>%
    group_by(year)%>%
    summarize(medianLifeExp=median(lifeExp),maxGdpPercap=max(gdpPercap))
ggplot(by_year,aes(x=year,y=medianLifeExp))+geom_point()+expand_limits(y=0)
```

```r
by_continent<-gapminder%>%
    group_by(continent,year)%>%
    summarize(medianLifeExp=median(lifeExp),maxGdpPercap=max(gdpPercap))
ggplot(by_continent,aes(x=year,y=medianLifeExp,color=continent))+geom_point()+expand_limits(y=0)
```
